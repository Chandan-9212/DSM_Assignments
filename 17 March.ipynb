Q1: What are missing values in a dataset? Why is it essential to handle missing values? Name some
algorithms that are not affected by missing values.

Missing values are the values that are not present in a dataset. They can be represented by NaN, NA, or any other placeholder. It is essential to handle missing values because they can affect the analysis and results of the model, leading to bias and incorrect predictions. Some algorithms that are not affected by missing values are tree-based algorithms like Random Forest and XGBoost.

Q3: Explain the imbalanced data. What will happen if imbalanced data is not handled?

Imbalanced data refers to a dataset where the distribution of the target variable is skewed towards one class. If imbalanced data is not handled, the machine learning model can be biased towards the majority class, resulting in poor predictions for the minority class. This can be detrimental in applications like fraud detection or disease diagnosis, where the cost of false negatives is high.

Q4: What are Up-sampling and Down-sampling? Explain with an example when up-sampling and down-
sampling are required.

Up-sampling involves duplicating the minority class samples to increase their representation in the dataset.
Example: Using the resample() function in pandas to up-sample the minority class.

Down-sampling involves reducing the majority class samples to balance the distribution of the target variable.
Example: Using the resample() function in pandas to down-sample the majority class.

Q5: What is data Augmentation? Explain SMOTE.

Data Augmentation is a technique used in machine learning to increase the amount of data available for training by creating new data based on the existing data. It involves applying various transformations to the existing data, such as rotating, scaling, or cropping images, adding noise or distortions, and changing the brightness or contrast. SMOTE, or Synthetic Minority Over-sampling Technique, is a popular data augmentation method used to address imbalanced datasets. It creates synthetic samples of the minority class by interpolating between the nearest neighbors of the minority class, thereby increasing the number of samples in the minority class.

Q6: What are outliers in a dataset? Why is it essential to handle outliers?

Outliers are data points that lie significantly outside the majority of the data points in a dataset. They can arise due to measurement errors, data processing errors, or genuine extreme values. Outliers can have a significant impact on the statistical analysis and machine learning models, leading to incorrect results and biased models. It is essential to handle outliers by detecting and removing or correcting them, depending on the nature of the data and the analysis.

Q7: You are working on a project that requires analyzing customer data. However, you notice that some of
the data is missing. What are some techniques you can use to handle the missing data in your analysis?

Deleting the rows or columns with missing data, if the amount of missing data is small and does not significantly affect the analysis.
Imputing the missing data with a suitable value, such as the mean, median, or mode of the available data.
Using advanced imputation methods, such as k-nearest neighbors or multiple imputation, which take into account the relationships between the variables and the patterns of missing data.

Q8: You are working with a large dataset and find that a small percentage of the data is missing. What are
some strategies you can use to determine if the missing data is missing at random or if there is a pattern
to the missing data?

Conducting a Missing Completely At Random (MCAR) test to determine if the missing data is unrelated to the other variables in the dataset.
Conducting a Missing At Random (MAR) test to determine if the missing data is related to some of the other variables in the dataset.
Conducting a Missing Not At Random (MNAR) test to determine if the missing data is related to the missing variable itself.

Using evaluation metrics that are not affected by class imbalance, such as the Area Under the ROC Curve (AUC-ROC) or Precision-Recall Curve.
Using sampling techniques, such as oversampling the minority class or undersampling the majority class, to balance the dataset before training the model.
Using cost-sensitive learning, where the misclassification cost of the minority class is given more weight than the majority class.

Q10: When attempting to estimate customer satisfaction for a project, you discover that the dataset is
unbalanced, with the bulk of customers reporting being satisfied. What methods can you employ to
balance the dataset and down-sample the majority class?

Random under-sampling, where a random subset of the majority class is selected to match the size of the minority class.
Tomek Links, where pairs of data points from different classes that are nearest neighbors are identified and the majority class points are removed.
Cluster Centroids, where the majority class data points are replaced with the centroids of K-means clusters of the majority class.

balance the dataset and up-sample the minority class?

Random over-sampling, where new samples are created by duplicating existing samples from the minority class.
