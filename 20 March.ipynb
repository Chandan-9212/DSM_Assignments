```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['sex']])
```

```python
encoded = encoded.toarray()
```

```python
encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoded
```

```python
pd.concat([df,encoded],axis=1)
```

A1. Data encoding is the process of converting data from one form to another, typically from a human-readable format to a machine-readable format. In data science, encoding is useful for transforming categorical data into numerical data, which can be used in machine learning algorithms.

A2. Nominal encoding is a technique used to encode categorical data where each category is assigned a unique integer value. For example, in a real-world scenario, we can use nominal encoding to encode the different types of cars available in a car showroom. Suppose we have three types of cars: Sedan, SUV, and Hatchback. We can assign integer values such as 1, 2, and 3 to these categories, respectively.

A3. Nominal encoding is preferred over one-hot encoding when the number of unique categories is relatively small. One practical example is encoding the seasons of the year, which have only four categories: Winter, Spring, Summer, and Fall. Using one-hot encoding for this scenario would create four new columns, each representing a season, which could result in overfitting in machine learning models.

A4. If the categorical data has only five unique values, we can use nominal encoding to transform the data into a format suitable for machine learning algorithms. Nominal encoding assigns a unique integer value to each category, and this technique works well for small numbers of unique categories.

A5. If we were to use nominal encoding to transform two categorical columns, we would create two new columns. Since nominal encoding assigns a unique integer value to each category, each column would be replaced with a new column containing integer values.

A6. For categorical data containing information about different types of animals, we can use one-hot encoding to transform the data into a format suitable for machine learning algorithms. One-hot encoding creates a new column for each unique category and assigns a binary value to each row indicating the presence or absence of the category.

To transform the categorical data into numerical data for the customer churn dataset, we can use one-hot encoding for the gender and contract type columns since they have a small number of unique categories.
