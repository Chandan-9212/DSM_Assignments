**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.**

import numpy as np

np.random.seed(3)


x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:**

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

**Ans.** The subplot() function in Matplotlib is used to create multiple plots within a single figure. It takes three arguments: the number of rows, the number of columns, and the plot number.

The subplot() function allows you to compare and analyze multiple plots side-by-side, making it a useful tool for data visualization.

Here's an example of how you can use the subplot() function to create four line plots using the data provided:

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10,10))

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.show()
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

import numpy as np

**Ans.** A bar plot, also known as a bar chart, is a type of graph that displays data as rectangular bars with heights or lengths proportional to the values they represent. The bars can be oriented horizontally or vertically, and the length or height of each bar represents a specific data value or range.

Bar plots are used to display and compare the values of different categories or groups of data. They are commonly used to visualize categorical data, such as frequency counts, survey results, or sales figures, and are especially useful for showing the differences or similarities between different groups.

Bar plots are useful because they allow you to quickly and easily compare the values of different categories or groups of data. They are also easy to read and interpret, making them an effective tool for communicating complex information to others.

In addition to traditional bar plots, there are many variations and types of bar plots, such as stacked bar plots, grouped bar plots, and percent stacked bar plots, which allow you to display additional information or multiple variables within the same graph.

Overall, bar plots are a powerful and versatile tool for data visualization and analysis that can be used in a variety of fields, including business, science, engineering, and social sciences.
